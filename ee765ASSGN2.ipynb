```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
from scipy.stats import norm
from math import exp


# /// Question 2
def pdf(t):
 return 1/(1+t)**2

def cdf(t):
  return -1/(1+t)

def R(t):
  return (t+2)/(t+1)
def fr(t):
  return 1/(t**2 + 3*t +2)

t= np.linspace(-3,1,200)
fig1 = plt.figure(figsize = (10, 5))
fig1.text(0.9, 0.15, 'Niraj kumar', 
         fontsize = 12, color ='green',
         ha ='right', va ='bottom', 
         alpha = 0.7)

plt.plot(t,pdf(t), label='PDF')
plt.plot(t,cdf(t), label='CDF')
plt.plot(t,R(t), label='Reliability')
plt.plot(t,fr(t), label='Failure rate')
plt.legend()
plt.xlabel('time (t)')
plt.ylabel('pdf,cdf,Reliability,failure rate')
plt.ylim([-200, 300])
plt.show()

# use """ ----  """ to comment other question part and run one question at a time
# although all code can be run at once
# /// question 3

def pdfQ3(t):
  return (100*t)*np.exp(-50*(t**2))

def frQ3(t):
   return 100*t

def RQ3(t):
  return np.exp(-50* t**2)  
def cdfQ3(t):
  return 1-(np.exp(-50* t**2))


t= np.linspace(0,0.5,200)
fig2 = plt.figure(figsize = (10, 5))
fig2.text(0.9, 0.15, 'Niraj kumar', 
         fontsize = 12, color ='green',
         ha ='right', va ='bottom', 
         alpha = 0.7)

plt.plot(t,pdfQ3(t), label='PDF')
plt.plot(t,cdfQ3(t), label='CDF')
plt.plot(t,RQ3(t), label='Reliability')
plt.plot(t,frQ3(t), label='Failure rate')
plt.legend()
plt.ylim([-10, 10])
plt.xlabel('time (t)')
plt.ylabel('pdf,cdf,Reliability,failure rate')
plt.show()



# Question 4
# PART A
CR= np.linspace(0,20,300)

pdfQ4= norm.pdf(CR,loc=10,scale=2)
def cdfQ4(CR):
  return norm.cdf(CR,loc=10,scale=2)


fig3 = plt.figure(figsize = (10, 5))
plt.plot(CR,pdfQ4,label='PDF')
plt.plot(CR,cdfQ4(CR),label='CDF')

plt.legend(loc='upper right')
plt.xlabel('CURRENT RATING')
plt.ylabel('PDF , CDF ')
plt.show()

#PART B
yield_B= quad(cdfQ4, 9, 11)

yield_total=quad(cdfQ4, 0, 20)
print(yield_B,yield_total)
fig4 = plt.figure(figsize = (10, 5))
plt.plot(CR,cdfQ4(CR),label='CDF')
plt.plot(CR,pdfQ4,label='PDF')
plt.fill_between(CR, cdfQ4(CR), where=[(CR > 9) and (CR < 11) for CR in CR], color='green',alpha=0.5,label= yield_B)
plt.legend(loc='upper right')
plt.xlabel('CURRENT RATING')
plt.ylabel(' CDF ')
plt.show()

#part C
yield_C= quad(cdfQ4,7,12)
fig5 = plt.figure(figsize = (10, 5))
plt.plot(CR,cdfQ4(CR),label='CDF')
plt.plot(CR,pdfQ4,label='PDF')
plt.fill_between(CR, cdfQ4(CR), where=[(CR > 7) and (CR < 12) for CR in CR], color='red',alpha=0.5,label= yield_C)
plt.legend(loc='upper right')
plt.xlabel('CURRENT RATING')
plt.ylabel(' CDF ')
plt.show()

# PART D
yield_90= quad(cdfQ4,1,19)
fig6 = plt.figure(figsize = (10, 5))
plt.plot(CR,cdfQ4(CR),label='CDF')
plt.plot(CR,pdfQ4,label='PDF')
plt.fill_between(CR, cdfQ4(CR), where=[(CR > 1) and (CR < 19) for CR in CR], color='yellow',alpha=0.5,label= yield_90)
plt.legend(loc='upper right')
plt.xlabel('CURRENT RATING')
plt.ylabel(' CDF ')
plt.show()
```
